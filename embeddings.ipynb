```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Artificial Intelligence is transforming the world.",
    "Coding is a valuable skill in today's world.",
    "Practice makes perfect, especially in programming.",
    "Always strive for continuous learning and growth."
]
```

```python
voc_size=10000
```

#### one hot Representation

```python
one_hot_repr=[one_hot(sent,voc_size) for sent in sentences]
one_hot_repr
```

#### Now convert into word embeddings

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences
import numpy as np
```

To make sentence of equal dimension, we use pad_sequence

```python
max_len=10
Embedding_docs= pad_sequences(one_hot_repr,padding='post',maxlen=max_len)
Embedding_docs
```

```python
dim=20
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=max_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(Embedding_docs)
```
